# ELE8812 - Traitement fréquentiel

## Démo : Filtrage passe-bas, filtres classiques

### Référence : DIP 4e édition, Chapitre 4, Section 4.8

Dans ce démo, une image sera filtrée dans le domaine fréquentiel en utilisant 3 différents types de filtres : filtre idéal passe-bas, gaussien passe-bas et Butterworth passe-bas. Pour ce faire, la transformée de Fourier (TF) de l’image sera d’abord effectuée. Le spectre de fréquences de l’image sera multiplié élément-par-élément par la réponse fréquentielle des filtres. Une TF inverse sera ensuite appliquée avant de pouvoir visualiser l’image filtrée. Trois paramètres de filtrage pourront être modifiés.

```python
# Charger image
A = plt.imread('Repo/Data/Barbara.tif');
plt.imshow(A, cmap='gray'); plt.title('Image Originale'); plt.axis('off'); plt.show();
```

```python
def filtpb_butter(A, fc, ordre, pad):
    """ Application d'un filtre Butterworth sur une image im à une fréquence de coupure fc dans 
        le domaine fréquentiel
    
    A (ndarray): Image d'entrée
    fc (float): Fréquence de coupure
    ordre (int): Ordre du filtre
    pad (int): Nombre de pixels de padding
    """
    M,N = A.shape
    M+=pad
    N+=pad

    # Image dans le domaine fréquentiel
    AA = np.fft.fftshift(np.fft.fft2(A, (M,N)))

    M0 = np.ceil((M+1) / 2)
    N0 = np.ceil((N+1) / 2)

    U, V = np.mgrid[1:M+1, 1:N+1]
    D2 = (U - M0)**2 + (V - N0)**2

    # Réponse fréquentielle du filtre Butterworth
    HH = 1 / (1 + (D2 / fc**2)**ordre)
    
    # Application du filtre et retour au domaine spatial
    BB = np.fft.ifftshift(AA * HH)
    B = np.fft.ifft2(BB)
    B = np.real(B)
    return B[:A.shape[0],:A.shape[1]] 
```

```python
def filtpb_gauss(A, fc, pad):
    """ Application d'un filtre gaussien sur une image im à une fréquence de coupure fc dans 
        le domaine fréquentiel
    
    A (ndarray): Image d'entrée
    fc (float): Fréquence de coupure
    pad (int): Nombre de pixels de padding
    """
    M,N = A.shape
    M+=pad
    N+=pad

    # Image dans le domaine fréquentiel
    AA = np.fft.fftshift(np.fft.fft2(A, (M,N)))

    M0 = np.ceil((M+1) / 2)
    N0 = np.ceil((N+1) / 2)

    U, V = np.mgrid[1:M+1, 1:N+1]
    D2 = (U - M0)**2 + (V - N0)**2
    
    # Réponse fréquentielle du filtre gaussien
    HH = np.exp(-D2 / (2 * fc**2))
    
    # Application du filtre et retour au domaine spatial
    BB = np.fft.ifftshift(AA * HH)
    B = np.fft.ifft2(BB)
    B = np.real(B)
    return B[:A.shape[0],:A.shape[1]]       
```

```python
def filtpb_idéal(A, fc, pad):
    """ Application d'un filtre gaussien sur une image im à une fréquence de coupure fc dans 
        le domaine fréquentiel
    
    A (ndarray): Image d'entrée
    fc (float): Fréquence de coupure
    pad (int): Nombre de pixels de padding
    """
    M,N = A.shape
    M+=pad
    N+=pad

    # Image dans le domaine fréquentiel
    AA = np.fft.fftshift(np.fft.fft2(A, (M,N)))

    M0 = np.ceil((M+1) / 2)
    N0 = np.ceil((N+1) / 2)

    U, V = np.mgrid[1:M+1, 1:N+1]
    D2 = (U - M0)**2 + (V - N0)**2
    
    # Réponse fréquentielle du filtre idéal
    HH = (D2<= fc**2)
    
    # Application du filtre et retour au domaine spatial
    BB = np.fft.ifftshift(AA * HH)
    B = np.fft.ifft2(BB)
    B = np.real(B)
    return B[:A.shape[0],:A.shape[1]]     
```

## Fréquence de coupure

Pour les 3 types de filtres passe-bas, plus la fréquence de coupure diminue, plus les contours de l’image deviennent flous.  

```python
slider_fc = widgets.IntSlider(min=1, max=100, step=1, description='fc:', readout=True,)
display(slider_fc)
```

## Taille du prolongement de l'image

Cette variable contrôle le nombre de zéros ajoutés en périphérie de l’image avant d’effectuer la TF. L’équivalent dans le domaine spatial d’une multiplication dans le domaine fréquentiel est une convolution circulaire (hypothèse implicite de périodicité). On peut imaginer que l’image se répète à l’infini dans chaque direction. L’ajout de zéros avant le filtrage permet d’éviter que le résultat du filtrage aux bords de l’image soit affecté par les pixels du bord opposé. Sans prolongement (`pad`=0), cet effet de *wrapping* est légèrement visible à l’extrémité supérieure de l’image à gauche de la chaise. Le prolongement par des zéros diminue cet effet et introduit une bordure de faible intensité aux bords de l’image.

```python
slider_pad = widgets.IntSlider(min=0, max=10, step=1, description='pad:', readout=True,)
display(slider_pad)
```

## Ordre du filtre

Cette variable influence uniquement le filtre Butterworth. Lorsque l’ordre du filtre augmente, l’allure du filtre se rapproche d’un filtre idéal. La précision de coupure est meilleure mais on voit apparaitre des artéfacts d’oscillations (comme pour le filtre idéal).

```python
slider_order = widgets.IntSlider(min=1, max=8, step=1, description='n:', readout=True,)
display(slider_order)
```

## Résultats

```python
plt.subplot(131); plt.imshow(filtpb_idéal(A,slider_fc.value,slider_pad.value), cmap='gray'); plt.axis('off'); plt.title('Filtrage idéal')
plt.subplot(132); plt.imshow(filtpb_gauss(A,slider_fc.value,slider_pad.value), cmap='gray'); plt.axis('off'); plt.title('Filtrage gaussien')
plt.subplot(133); plt.imshow(filtpb_butter(A, slider_fc.value, slider_order.value, slider_pad.value), cmap='gray'); plt.axis('off'); plt.title('Filtrage butterworth')
plt.show()
```

**Filtre idéal :** Le filtre passe-bas idéal entraîne des artéfacts d’oscillation et ce, même à une fréquence de coupure élevée. Rappel : une multiplication dans le domaine fréquentiel correspond à une convolution dans le domaine spatial, la représentation spatiale d’un filtre idéal contient des oscillations (fonction [jinc](https://fr.wikipedia.org/wiki/Fonction_sombrero)).

**Filtre gaussien :**
Le filtre gaussien n’entraîne pas d’artéfacts d’oscillation : sa représentation spatiale décroît de façon monotonique. Or, on peut observer que pour une fréquence de coupure donnée, l’image filtrée avec un filtre gaussien paraît d’avantage floue que celle filtrée avec un filtre Butterworth. Due à la décroissance moins abrupte de la représentation fréquentielle du filtre, la notion de fréquence de coupure est moins bien définie pour un filtre gaussien.

**Filtre Butterworth :** Comme mentionné plus haut, plus l’ordre d’un filtre Butterworth est élevé, plus sa représentation se rapproche d’un filtre idéal et plus il y a présence d’artéfacts d’oscillation dans l’image filtrée.

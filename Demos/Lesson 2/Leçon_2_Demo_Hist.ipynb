## Démo Histogramme : Relation entre l'aspect d'une image et son histogramme

## Référence : DIP 4e édition, Chapitre 3, p.134 Figure 3.16


L’histogramme d’une image permet de visualiser la distribution des valeurs d’intensité de ses pixels. L’axe des x correspond aux valeurs d’intensité possibles, allant de 0 à L-1. Pour chaque valeur d’intensité r<sub>k</sub>, le nombre de pixel ayant l’intensité r<sub>k</sub> y est associé.

Dans ce démo, 4 images ayant différentes caractéristiques d’intensité sont présentées : intensité faible, intensité élevée, faible contraste et contraste élevé. Ces caractéristiques peuvent être identifiées en observant les histogrammes des images.

```python
import numpy as np
import matplotlib.pyplot as plt

# Option d'affichage
plt.rcParams['figure.figsize'] = (16,9)
plt.rcParams['axes.linewidth'] = 3
plt.rcParams['axes.titlesize'] = 15
plt.rcParams['axes.titlepad'] = 15
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

```python
# Charger images
A = plt.imread('Repo/Data/Fig0316(1)(top_left).tif')
B = plt.imread('Repo/Data/Fig0316(2)(2nd_from_top).tif');
C = plt.imread('Repo/Data/Fig0316(3)(third_from_top).tif');
D = plt.imread('Repo/Data/Fig0316(4)(bottom_left).tif');
```

## Exemple 1

Le premier exemple présente une image à haute intensité. La distribution des pixel est biaisée vers la droite, c’est-à-dire vers les valeurs d’intensité élevée.

```python
plt.subplot(121); plt.imshow(A, cmap='gray', vmin=0, vmax=255); plt.colorbar(); plt.title('Image 1')
plt.subplot(122); plt.hist(A.ravel(),256,[0,256]); plt.title('Histogramme 1')
plt.show()  
```

## Exemple 2

Le deuxième exemple présente une image à faible contraste. L’image à l’apparence délavée ne contient pas de pixels à très faible ou très haute intensité. Ainsi, l’histogramme d’une telle image est habituellement étroit et situé au centre de l’échelle d’intensité. 

```python
plt.subplot(121); plt.imshow(B, cmap='gray', vmin=0, vmax=255); plt.colorbar(); plt.title('Image 1')
plt.subplot(122); plt.hist(B.ravel(),256,[0,256]); plt.title('Histogramme 1')
plt.show()  
```

## Exemple 3

Le troisième exemple présente une image à contraste élevé. Pour cette image, l’histogramme qui s’étend de la valeur d’intensité la plus faible (0) à la valeur la plus élevée (255) couvre une grande gamme de valeur d’intensité. De plus, sa distribution de valeurs d’intensité est plus proche d’une distribution uniforme que les exemples précédents.

```python
plt.subplot(121); plt.imshow(C, cmap='gray', vmin=0, vmax=255); plt.colorbar(); plt.title('Image 1')
plt.subplot(122); plt.hist(C.ravel(),256,[0,256]); plt.title('Histogramme 1')
plt.show()  
```

## Exemple 4

Le quatrième exemple présente une image à faible intensité, d’apparence sombre. La distribution des pixels est concentrée vers la gauche, c’est-à-dire vers les valeurs d’intensité faible.

```python
plt.subplot(121); plt.imshow(D, cmap='gray', vmin=0, vmax=255); plt.colorbar(); plt.title('Image 1')
plt.subplot(122); plt.hist(D.ravel(),256,[0,256]); plt.title('Histogramme 1')
plt.show()  
```

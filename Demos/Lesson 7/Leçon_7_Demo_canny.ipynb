## Démo :  Détection de contours par la méthode de Canny

## Détection de contour par méthode Canny

```python
# Charger image
A = plt.imread('Repo/Data/cerveau.tif');
plt.imshow(A, cmap='gray'); plt.title('Image Originale'); plt.axis('off'); plt.show();
```

```python
# Valeurs par défaut :
# contours = binary image with 1s where there is an edge and 0s elsewhere
# sigma = standard deviation of Gaussian
# If you do not specify thresholds, then edge chooses the value or values automatically.
seuil = [5e-02, 0.125]
sigma = 2

contours = canny(A, sigma=sigma, low_threshold=None, high_threshold=None)

fac = 3
imc = np.minimum(np.double(A)/(256*fac-1) + contours, 1)

plt.imshow(imc, cmap='gray'); plt.title('Canny, \u03C3 = '+str(sigma)+', seuil par défaut'); plt.axis('off'); plt.show();
```

## Effet de sigma

```python
# paramètres 1
seuil = [5e-02, 0.125]
sigma = 2

contours = canny(A, sigma=sigma, low_threshold=None, high_threshold=None)

fac = 3
imc = np.minimum(np.double(A)/(256*fac-1) + contours, 1)

plt.subplot(121)
plt.imshow(imc, cmap='gray'); plt.title('Canny, \u03C3 = '+str(sigma)+', seuil par défaut'); plt.axis('off');


# paramètres 2
seuil = [5e-02, 0.125]
sigma = 4

contours = canny(A, sigma=sigma, low_threshold=None, high_threshold=None)

fac = 3
imc = np.minimum(np.double(A)/(256*fac-1) + contours, 1)

plt.subplot(122)
plt.imshow(imc, cmap='gray'); plt.title('Canny, \u03C3 = '+str(sigma)+', seuil par défaut'); plt.axis('off'); plt.show();
```
